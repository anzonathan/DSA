Sorting a random array into n random groups. 

```python
import random 

no = ['Nathaniel','Matthew','Richard','Daniel','Brenda','Xato','Arnold','James','Rock','Wallace','Anitah']

n = 3

g = []

for i in range(n):
    h = []
    for j in range(n):
        x = random.choice(no)
        h.append(x)
        no.remove(x)
    g.append(h)
    

print(g)
```
